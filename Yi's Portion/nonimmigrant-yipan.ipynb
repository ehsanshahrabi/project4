```python
# Load file

nonimmsuptable_state_paths = [
'NonImmigrant_Visa/fy2005_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2006_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2007_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2008_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2009_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2010_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2011_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2012_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2013_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2014_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2015_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2016_nonimmsuptable3d.xls',
'NonImmigrant_Visa/fy2017_nonimmsuptable3d.xlsx',
'NonImmigrant_Visa/fy2018_nonimmsuptable3d.xlsx',
'NonImmigrant_Visa/fy2019_nonimmsuptable3d.xlsx',
'NonImmigrant_Visa/fy2020_nonimmsuptable3d.xlsx',
'NonImmigrant_Visa/fy2021_nonimmsuptable3d.xlsx']

nonimmsuptable_state_c = []

for idx, nonimmsuptable_state_path in enumerate(nonimmsuptable_state_paths):

    # Read excel

    nonimmsuptable_state = pd.read_excel(nonimmsuptable_state_path, header=None)
    # clean the data
    nonimmsuptable_state = nonimmsuptable_state.drop(nonimmsuptable_state.index[0:3])
    nonimmsuptable_state = nonimmsuptable_state.reset_index(drop=True)
    nonimmsuptable_state = nonimmsuptable_state.set_axis(nonimmsuptable_state.iloc[0],axis = "columns")
    nonimmsuptable_state = nonimmsuptable_state[2:58]

    # Add Year column in the first row
    nonimmsuptable_state.insert(0, "Year", 2005 + idx)
    nonimmsuptable_state_c.append(nonimmsuptable_state)

result_df2 = pd.concat(nonimmsuptable_state_c, ignore_index=True)
result_df2.drop(columns='Total', inplace=True) 

result_df2
```

```python
nonimmsuptable_state = result_df2

nonimmsuptable_state.head()
```

```python
# Unpivot tables

nonimmsuptable_country = pd.melt(nonimmsuptable_country, id_vars = ["Year", "Class", "Description"], var_name = "country", value_name="population_count")
nonimmsuptable_state = pd.melt(nonimmsuptable_state, id_vars = ["Year","Class", "Description"], var_name = "state", value_name="population_count")

# Filter Visa Type
# nonimmsuptable_country = nonimmsuptable_country[nonimmsuptable_country["Class"].isin(["F1","F2","F3","H1B","H1B1","H4","K1","K2"])]
# nonimmsuptable_state = nonimmsuptable_state[nonimmsuptable_state["Class"].isin(["F1","F2","F3","H1B","H1B1","H4","K1","K2"])]

# Convert "D" with 0 in the value
nonimmsuptable_country["population_count"] = nonimmsuptable_country["population_count"].replace(["D","-"], 0)
nonimmsuptable_state["population_count"] = nonimmsuptable_state["population_count"].replace(["D","-"], 0)

# clean the class column
nonimmsuptable_country["Class"] = nonimmsuptable_country["Class"].str.replace(r'(\w)-(\d)', r'\1\2')
nonimmsuptable_state["Class"] = nonimmsuptable_state["Class"].str.replace(r'(\w)-(\d)', r'\1\2')
```
